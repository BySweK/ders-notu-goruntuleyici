## Kullanıcıdan alınan veriler ile ona özel grafik olusturma

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Kullanıcıdan Veriler Alınır

dersler = []
notlar = []

while(True):
 ders = input("Ders adi gir:")
 dersNot = int(input("Ders notu gir:"))
 dersler.append(ders)
 notlar.append(dersNot)
 devamSarti = int(input("Devam etmek istiyor musun ? 1-Evet, 2-Hayir"))
 if devamSarti == 2:
    break

# Verileri tutan listeler Numpy array'e cevirilir.
dersler = np.array(dersler)
notlar = np.array(notlar)

# Grafik cizilir

plt.plot(dersler,notlar,"g*-")
plt.xlabel("Dersler")
plt.ylabel("Notlar")


               

               
```
